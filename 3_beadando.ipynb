```python
#Importálás
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Adatbeolvasás
data = pd.read_csv('heart_attack_prediction_dataset.csv')
data.head()
```

```python
#Hány országból vannak páciensek?
print(f"{len(data['Country'].unique())} országból vannak páciensek")
```

```python
#Egy országra átlagban, mediánban hány páciens jut és mi ennek a szórása?
table = data.groupby('Country')['Patient ID'].count()
table
```

```python
avg = np.mean(table)
med = np.median(table)
szor = np.std(table)
print(f"Átlag: {avg}")
print(f"Medián: {med}")
print(f"Szórás: {np.round(szor, 2)}")
```
